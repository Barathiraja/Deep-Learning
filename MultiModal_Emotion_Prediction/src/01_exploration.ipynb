This notebook explores the WESAD dataset structure, inspects signals, sampling rates, and prepares for alignment.

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Modify this path if your dataset is elsewhere
DATA_PATH = 'D:\\Dissertation\\Data\\WESAD\\S2\\S2.pkl'

with open(DATA_PATH, 'rb') as f:
    data = pickle.load(f, encoding='latin1')

print('Top-level keys:', data.keys())
print('Signal keys:', data['signal'].keys())
print('Label shape:', data['label'].shape)
```

```python
# Inspect chest & wrist sensors
chest = data['signal']['chest']
wrist = data['signal']['wrist']

print('Chest sensors:', chest.keys())
print('Wrist sensors:', wrist.keys())
```

```python
# Print individual signal shapes
for key, val in chest.items():
    print(f'Chest {key}: {np.array(val).shape}')

for key, val in wrist.items():
    print(f'Wrist {key}: {np.array(val).shape}')
```
